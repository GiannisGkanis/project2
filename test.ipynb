```python
from dash import Dash
from dash import html, dcc
from dash.dependencies import Input, Output
```

```python
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
```

```python
app = Dash(external_stylesheets=external_stylesheets)
app.layout = html.Div([
    html.H1('Bangalore'),
    #dcc.Graph(id= 'grahp1',figure = {},style={'width': '48%', 'display': 'inline-block'}),
    #dcc.Graph(id= 'grahp2',figure = {},style={'width': '48%', 'display': 'inline-block'}),
    #dcc.Graph(id= 'grahp3',figure = {},style={'width': '48%', 'display': 'inline-block','height': '400px'}),
    html.Iframe(id='map', srcDoc=open('map1.html', 'r').read(),
                style={'width': '45%', 'height': '300px','display': 'inline-block'}),
    
    dcc.Dropdown(df['Gender'].unique(),value='Male', id= 'dropdown1',style={'margin-top': '20px'} )
])

@app.callback(Output('map', 'srcDoc'),
              Input('dropdown1', 'value'))

def gender(x):
    df_temp = df[df['Gender'] == x]

    map1=folium.Map(location=[12.9766,77.5993],zoom_start=10.5)
    prices=[{"price":df_temp.iloc[i]['Gender'],"location":[df_temp.iloc[i]['latitude'],df_temp.iloc[i]['longitude']]} for i in range(df_temp.shape[0])]
    
    for price in prices:
        folium.Marker(location=price['location'],popup=price['price']).add_to(map1)
    map_html = map1.get_root().render()
    
    return map_html
```

```python
app.run(host='localhost', mode = 'inline', port= 8090, debug=False)
```
